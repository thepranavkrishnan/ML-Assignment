```python
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Define the path to your CSV file in Google Drive
file_path = '/content/drive/MyDrive/Machine learning college/WineQT.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()

```

```python
#A1
import numpy as np

# Assuming 'df' is your DataFrame containing the dataset
# Filter the data for quality ratings 5 and 6
class_5_data = df[df['quality'] == 5]
class_6_data = df[df['quality'] == 6]

# Calculate the mean (centroid) for each class
mean_5 = np.mean(class_5_data.drop(columns=['quality', 'Id']), axis=0)
mean_6 = np.mean(class_6_data.drop(columns=['quality', 'Id']), axis=0)

# Calculate the spread (standard deviation) for each class
std_5 = np.std(class_5_data.drop(columns=['quality', 'Id']), axis=0)
std_6 = np.std(class_6_data.drop(columns=['quality', 'Id']), axis=0)

# Calculate the distance between mean vectors between classes
distance_between_classes = np.linalg.norm(mean_5 - mean_6)

# Display the filtered data for each class
print("Class 5 Data:")
print(class_5_data.head())

print("\nClass 6 Data:")
print(class_6_data.head())

# Display the mean and standard deviation vectors for each class
print("\nMean (Centroid) for Class 5:")
print(mean_5)

print("\nMean (Centroid) for Class 6:")
print(mean_6)

print("\nSpread (Standard Deviation) for Class 5:")
print(std_5)

print("\nSpread (Standard Deviation) for Class 6:")
print(std_6)

# Display the distance between mean vectors
print("\nDistance between Mean Vectors (Class 5 and Class 6):", distance_between_classes)

```

```python
#A2
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the dataset
# Extract the 'fixed acidity' column data
fixed_acidity_data = df['fixed acidity']

# Generate a histogram using specified buckets (data ranges)
# You can adjust the number of bins (buckets) as needed
hist, bins = np.histogram(fixed_acidity_data, bins=20)  # Adjust the number of bins as needed

# Calculate the mean and variance of the 'fixed acidity' data
mean_fixed_acidity = np.mean(fixed_acidity_data)
variance_fixed_acidity = np.var(fixed_acidity_data)

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(fixed_acidity_data, bins=bins, edgecolor='k')
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.title('Histogram of Fixed Acidity')
plt.grid(True)

# Display the mean and variance
print("Mean Fixed Acidity:", mean_fixed_acidity)
print("Variance of Fixed Acidity:", variance_fixed_acidity)

# Show the histogram plot
plt.show()
```

```python
#A3
# Assuming 'df' is your DataFrame containing the dataset
# Extract the feature vectors for "fixed acidity" and "volatile acidity"
fixed_acidity_vector = df['fixed acidity'].values
volatile_acidity_vector = df['volatile acidity'].values

# Define a range of values for r from 1 to 10
r_values = np.arange(1, 11)

# Calculate the Minkowski distance for each value of r
minkowski_distances = []
for r in r_values:
    distance = np.linalg.norm(fixed_acidity_vector - volatile_acidity_vector, ord=r)
    minkowski_distances.append(distance)

# Create a plot to observe the nature of the graph
plt.figure(figsize=(10, 6))
plt.plot(r_values, minkowski_distances, marker='o', linestyle='-')
plt.xlabel('r (Minkowski Parameter)')
plt.ylabel('Minkowski Distance')
plt.title('Minkowski Distance vs. r')
plt.grid(True)

# Show the plot
plt.show()
```

```python
#A4
from sklearn.model_selection import train_test_split

# Define the feature matrix (X) and class labels (y) based on your dataset
X = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
y = df['quality']

# Split the dataset into a training set (70%) and a test set (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# The 'test_size' parameter specifies the proportion of the dataset to include in the test split.
# The 'random_state' parameter is used to ensure reproducibility; you can set it to any integer.

# Print the shapes of the resulting sets to verify the split
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#A5
from sklearn.neighbors import KNeighborsClassifier

# Assuming you have already defined X_train and y_train from the previous exercise
# If not, make sure to define X_train and y_train using your dataset

# Create a k-NN classifier with k = 3
neigh = KNeighborsClassifier(n_neighbors=3)

# Fit the classifier using the training data
neigh.fit(X_train, y_train)
```

```python
#A6
# Assuming you have already trained the k-NN classifier 'neigh' and defined X_test and y_test

# Test the accuracy of the k-NN classifier using the test set
accuracy = neigh.score(X_test, y_test)

# Print the accuracy
print("Accuracy of the k-NN classifier:", accuracy)
```

```python
#A7
# Assuming you have already trained the k-NN classifier 'neigh' and defined X_test

# Use the predict function to obtain predicted class labels for the test data
predicted_labels = neigh.predict(X_test)

# Print the predicted class labels
print("Predicted Class Labels for Test Data:")
print(predicted_labels)
```

```python
#A8
# Assuming you have already defined X_train, X_test, y_train, and y_test

# Create lists to store accuracy values for NN (k=1) and k-NN (k from 1 to 11)
nn_accuracies = []
knn_accuracies = []

# Loop through k values from 1 to 11
for k in range(1, 12):
    # Create and train the NN classifier with k = 1
    nn_classifier = KNeighborsClassifier(n_neighbors=1)
    nn_classifier.fit(X_train, y_train)

    # Create and train the k-NN classifier with the current k
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(X_train, y_train)

    # Calculate accuracy for NN
    nn_accuracy = nn_classifier.score(X_test, y_test)

    # Calculate accuracy for k-NN with the current k
    knn_accuracy = knn_classifier.score(X_test, y_test)

    # Append accuracies to the lists
    nn_accuracies.append(nn_accuracy)
    knn_accuracies.append(knn_accuracy)

# Plot the accuracy values for NN and k-NN
plt.figure(figsize=(10, 6))
plt.plot(range(1, 12), nn_accuracies, label='NN (k=1)', marker='o', linestyle='-')
plt.plot(range(1, 12), knn_accuracies, label='k-NN (k from 1 to 11)', marker='o', linestyle='-')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison: NN vs. k-NN (k from 1 to 11)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Assuming you have already defined X_train, X_test, y_train, and y_test
# Assuming k-NN classifier is trained and tested

# Create and train the k-NN classifier (you can choose an appropriate k value)
knn_classifier = KNeighborsClassifier(n_neighbors=3)
knn_classifier.fit(X_train, y_train)

# Predict class labels for training and test data
y_train_pred = knn_classifier.predict(X_train)
y_test_pred = knn_classifier.predict(X_test)

# Calculate confusion matrices for training and test data
confusion_matrix_train = confusion_matrix(y_train, y_train_pred)
confusion_matrix_test = confusion_matrix(y_test, y_test_pred)

# Generate classification reports for training and test data
report_train = classification_report(y_train, y_train_pred)
report_test = classification_report(y_test, y_test_pred)

# Print confusion matrices and classification reports
print("Confusion Matrix (Training Data):\n", confusion_matrix_train)
print("\nConfusion Matrix (Test Data):\n", confusion_matrix_test)

print("\nClassification Report (Training Data):\n", report_train)
print("\nClassification Report (Test Data):\n", report_test)
```
